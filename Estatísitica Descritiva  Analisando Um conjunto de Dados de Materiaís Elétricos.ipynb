
Quando falamos de análise de dados, abrimos as portas para um universo de inúmeras possibilidades. No entanto, para alcançar resultados significativos, é fundamental entender o problema em questão e o contexto dos dados disponíveis. Compreender a variável de interesse (y) e como a análise dessa variável pode gerar insights valiosos para a tomada de decisões é um passo essencial.

Uma das primeiras abordagens para esse processo é a **Estatística Descritiva**, que nos permite resumir e interpretar dados de maneira clara e objetiva. Vamos explorar como essa técnica pode ser usada para transformar dados em informações relevantes.

```python

# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando uma base de dados fictícia de vendas de equipamentos elétricos
data = {
    'Mes': ['Jan', 'Feb', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
    'Equipamento': ['Transformador', 'Gerador', 'Bateria', 'Inversor', 'Disjuntor', 'Transformador',
                    'Gerador', 'Bateria', 'Inversor', 'Disjuntor', 'Transformador', 'Gerador'],
    'Vendas': [5000, 6000, 7000, 7500, 8000, 6500, 7000, 7200, 7300, 7800, 8200, 7500]}
```

```python
df = pd.DataFrame(data)

# Exibindo o DataFrame
print("Base de Dados:")
print(df)
```


Levando em consideração que a Estatística Descritiva organiza, resume e apresenta os dados de forma compreensível, podemos afirmar que seu foco é fornecer uma visão clara das principais características de um conjunto de dados. Dentro desse contexto, podemos explorar as Medidas de Tendência Central, que nos ajudam a entender o comportamento central dos dados:



*   **Média**: A soma de todos os valores dividida pelo número total de observações.

*   **Mediana**: O valor central quando os dados estão ordenados, representando o ponto onde metade dos valores está acima e metade está abaixo.
*   **Moda**: O valor que ocorre com maior frequência no conjunto de dados.













```python
# Medidas de Tendência Central
media = df['Vendas'].mean()
mediana = df['Vendas'].median()
moda = df['Vendas'].mode()[0]

# Resultados
print(f'Média das Vendas: {media}')
print(f'Mediana das Vendas: {mediana}')
print(f'Moda das Vendas: {moda}')
```

A identificação de tendências nas vendas é crucial para entender o desempenho geral dos produtos e tomar decisões estratégicas baseadas em dados. Usar medidas como a média e a mediana ajuda a perceber padrões no comportamento das vendas ao longo do tempo ou por diferentes categorias de produtos.

```python
# Análise das Vendas Mensais com Medidas de Tendência Central
bars = plt.bar(df['Mes'], df['Vendas'], color='skyblue', edgecolor='black', label='Vendas Mensais')
plt.axhline(media, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media:.2f}')
plt.axhline(mediana, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana:.2f}')
plt.axhline(moda, color='b', linestyle='dashed', linewidth=2, label=f'Moda: {moda:.2f}')
plt.title('Vendas Mensais com Medidas de Tendência Central')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.legend()

# Definindo cores para barras
cores = ['orange' if venda < media and venda < mediana and venda < moda else 'skyblue' for venda in df['Vendas']]
for bar, cor in zip(bars, cores):
    bar.set_color(cor)

plt.tight_layout()
plt.show()
```

O que podemos interpretar na análise anterior usando Tendências Centrais é:




*   **Média (Linha Vermelha)**: Representa o valor médio das vendas mensais. Se alguns meses estão muito abaixo da média, pode indicar problemas pontuais, como falhas na cadeia de suprimentos ou marketing. **Ação**: Investigar e corrigir esses problemas pode melhorar o desempenho.
*   **Mediana (Linha Verde)**: Representa o ponto central das vendas, menos influenciado por valores extremos. Se a mediana for significativamente menor que a média, pode indicar distorção causada por vendas muito altas. **Ação**: Comparar média e mediana ajuda a identificar o desempenho típico.
*   **Moda (Linha Azul)**: Representa o valor de vendas mais frequente. Identificar meses com vendas recorrentes ajuda a detectar padrões sazonais. **Ação**: Alinhar estratégias de marketing e estoque com esses padrões melhora o planejamento.













```python
# Medidas de Tendência Central para Produtos
vendas_por_produto = df.groupby('Equipamento')['Vendas'].sum().sort_values()
media_produto = vendas_por_produto.mean()
mediana_produto = vendas_por_produto.median()
moda_produto = vendas_por_produto.mode()[0]

# Resultados
print(f'Média das Vendas por Produto: {media_produto}')
print(f'Mediana das Vendas por Produto: {mediana_produto}')
print(f'Moda das Vendas por Produto: {moda_produto}')
```

```python
# Análise de Vendas Totais por Produto
bars = plt.bar(vendas_por_produto.index, vendas_por_produto, color='skyblue', edgecolor='black', label='Vendas Totais por Produto')
plt.axhline(media_produto, color='r', linestyle='dashed', linewidth=2, label=f'Média: {media_produto:.2f}')
plt.axhline(mediana_produto, color='g', linestyle='dashed', linewidth=2, label=f'Mediana: {mediana_produto:.2f}')
plt.axhline(moda_produto, color='b', linestyle='dashed', linewidth=2, label=f'Moda: {moda_produto:.2f}')
plt.title('Vendas Totais por Produto com Medidas de Tendência Central')
plt.xlabel('Produto')
plt.ylabel('Vendas')
plt.legend()
plt.xticks(rotation=45, ha='right')

# Destacando as barras que ficam abaixo da média, mediana e moda
cores = ['orange' if venda < media_produto and venda < mediana_produto and venda < moda_produto else 'skyblue' for venda in vendas_por_produto]
for bar, cor in zip(bars, cores):
    bar.set_color(cor)

plt.tight_layout()
plt.show()
```

**Para concluir:**

Também podemos utilizar da mesma abordagem para análisar outras variaveis do subconjunto, como por exemplo, os Produtos. Logo podemos ver de modo mais claro como nosso processo esta se comportando, para então tomar um decisão mais direcionada. Será que é viável manter a comercialização da linha de baterias? Geradores parece ser uma linha de produtos mais atrativa, o que podemos fazer para maximixar suas vendas?
